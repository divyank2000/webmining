```python
import numpy as np
X=np.array([[5,3],[10,15],[15,12],[24,10],[30,30],[85,70],[71,80],[60,78],[70,55],[80,91],]) 
import matplotlib.pyplot as plt
labels=range(1,11)
plt.figure(figsize=(10,7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X[:,0],X[:,1],label='True Position')
for label, x,y in zip(labels,X[:,0],X[:,1]):
 plt.annotate(
 label,
 xy=(x,y),xytext=(-3,3),
 textcoords='offset points',ha='right',va='bottom')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
from matplotlib import pyplot as plt
linked=linkage(X, 'single')
labelList=range(1,11)
plt.figure(figsize=(10,7))
dendrogram(linked,
 orientation='top',
 labels=labelList,
 distance_sort='descending',
 show_leaf_counts=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
cluster.fit_predict(X)
plt.scatter(X[:,0],X[:,1],c=cluster.labels_,cmap='rainbow')
```

```python
import random
data=[]
for i in range(100):
 t1=round(random.random()*100)
 t2=round(random.random()*100)
 data.append([t1,t2])
dataX=np.array(data)
labels=range(1,101)
plt.figure(figsize=(10,7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(dataX[:,0],dataX[:,1],label='True Position')
for label, x,y in zip(labels,dataX[:,0],dataX[:,1]):
 plt.annotate(
 label,
 xy=(x,y),xytext=(-3,3),
 textcoords='offset points',ha='right',va='bottom')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
from matplotlib import pyplot as plt
linked=linkage(dataX, 'single')
labelList=range(1,101)
plt.figure(figsize=(30,30))
dendrogram(linked,
 orientation='top',
 labels=labelList,
 distance_sort='descending',
 show_leaf_counts=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=9,affinity='euclidean',linkage='ward')
cluster.fit_predict(dataX)
plt.scatter(dataX[:,0],dataX[:,1],c=cluster.labels_,cmap='rainbow')
```
